```python
#Project task
# Author: Andras Csullog

# Write a program called analysis.py that:  
#     1. Outputs a summary of each variable to a single text file,  
#     2. Saves a histogram of each variable to png files, and  
#     3. Outputs a scatter plot of each pair of variables.  
#     4. Performs any other analysis you think is appropriate

# Importing the libraries which being used for data analysis
import matplotlib.pyplot as plt    # creating the plots
import pandas as pd                # for reading the data file and manipulating the data
import seaborn as sns              # for creating staistical graphics
```

```python
# Reading the data CSV file with panda called pd - but it can be any name - and defining as 'irisData'
irisData = pd.read_csv('iris.csv')
```

```python
# Listing irisData and counting the rowsxcolumns
irisData
```

```python
# Outputting the SUMMARY of variables into a TXT file called summary.txt where the second "w" ensures that the lines overriding at each program running to get the latest state
output = open("summary.txt", "w")

# Printing out the summarized calculations from irisData and saving into output

output.write(" **********   Summary of the Iris Data set and its variables  **********   \n")
output.write("\n")

# using the head method for listing the first 5 rows from the data set and formatting
output.write("              Heading the first 5 lines of the Iris Data\n")
output.write("---------------------------------------------------------------------------\n")
print(irisData.head(), file = output)
output.write("---------------------------------------------------------------------------\n")

# using the sum() method for the data set and reusing the previous format
output.write("                    Sum the Iris Data\n")
output.write("---------------------------------------------------------------------------\n")
print(irisData.sum(), file = output)
output.write("---------------------------------------------------------------------------\n")

# using the cov() method for the data set and reusing the previous format
output.write("                    Covariance matrix of the Iris Data\n")
output.write("---------------------------------------------------------------------------\n")
print(irisData.cov(), file = output)
output.write("---------------------------------------------------------------------------\n")

# using the corr() method for the data set and reusing the previous format - for checking the relationship between the columns
output.write("                      Correlation Iris Data\n")
output.write("---------------------------------------------------------------------------\n")
print(irisData.corr(), file = output)
output.write("---------------------------------------------------------------------------\n")

# using the describe() method for the data set and reusing the previous format
output.write("                    Summarizing the Iris Data\n")
output.write("---------------------------------------------------------------------------\n")
print(irisData.describe(), file = output)
output.write("---------------------------------------------------------------------------\n")

species=irisData.groupby(['Species'])
output.write("                    Summarizing the Iris Data\n")
output.write("---------------------------------------------------------------------------\n")
print(species.describe(include = 'all'), file = output)
output.write("---------------------------------------------------------------------------\n")

# Discovering the duplicated() method for the data set and reusing the previous format
output.write("                    Duplicates the Iris Data\n")
output.write("---------------------------------------------------------------------------\n")
print(irisData.duplicated(), file = output)
output.write("---------------------------------------------------------------------------\n")

# Discovering the shape method for the data set and reusing the previous format - string it in tuple data type
output.write("                    Number of rows and columns the Iris Data\n")
output.write("---------------------------------------------------------------------------\n")
print(irisData.shape, file = output)
output.write("---------------------------------------------------------------------------\n")

# using the tail() method for listing the first 5 rows from the data set and formatting
output.write("              Printing the last 5 rows Iris Data\n")
output.write("---------------------------------------------------------------------------\n")
print(irisData.tail(), file = output)
output.write("---------------------------------------------------------------------------\n")

#Closing the output file when finished
output.close()

```

```python
# Defining SCATTER plot with x and y axis with Seaborn library as sns
sns.scatterplot(x = "SepalLength_Cm" , y = "SepalWidth_Cm", data = irisData, hue = "Species", palette = ['green', 'red', 'blue'])

# Creating a title for the plot
plt.title('Sepal Length and Width Scatter Plot')

# Saving the output result in PNG and store in the images folder
# For many times the plot could not be saved into PNG as it was after plt.show() and didin't work, so it works if placing before it
plt.savefig("images\scatter_plot.png")

# Displays the plot
plt.show()

sns.scatterplot(x = "PetalLength_Cm" , y = "PetalWidth_Cm", data = irisData, hue = "Species", palette = ['green', 'red', 'blue'])
plt.title('Petal Length and Width Scatter Plot')

# Saving the output result in PNG and store in the images folder
plt.savefig("images\scatter_plot2.png")

plt.show()        

sns.scatterplot(x = "SepalLength_Cm" , y = "PetalLength_Cm", data = irisData, hue = "Species", palette = ['green', 'red', 'blue'])
plt.title('Petal vs Sepal Length Scatter Plot')

# Saving the output result in PNG and store in the images folder
plt.savefig("images\scatter_plot3.png")

plt.show()

sns.scatterplot(x = "SepalWidth_Cm" , y = "PetalWidth_Cm", data = irisData, hue = "Species", palette = ['green', 'red', 'blue'])
plt.title('Petal vs Sepal Width Scatter Plot')

# Saving the output result in PNG and store in the images folder
plt.savefig("images\scatter_plot4.png")

plt.show()
```

```python
with sns.axes_style('white'):
    sns.jointplot(x="SepalLength_Cm", y="SepalWidth_Cm", data=irisData, kind='hex')
# Saving the output result in PNG and store in the images folder
plt.savefig("images\joint_plot.png")    
```

```python
# scatter plot based on the category, before I used swarm plot but it complaint as it will be deprecated
ax = sns.stripplot(x="SepalLength_Cm", y= "SepalWidth_Cm", data=irisData , hue = "Species")
# Saving the output result in PNG and store in the images folder
plt.savefig("images\striplot.png")
```

```python
# Plot pairwise relationships, added semantic mapping with ue variable and and changes the default marginal plot to a layered kernel density estimate:
sns.pairplot(data=irisData, hue="Species", palette=["green", "red", "blue"])
# Saving the output result in PNG and store in the images folder
plt.savefig("images\pair_plot.png")
```

```python
# histplot() to draw both bivariate and univariate histograms:
sns.pairplot(irisData, kind="hist")
plt.savefig("images\pair_hist.png")
```

```python
# Creating a histogram
sns.displot(irisData, x='SepalLength_Cm', col='Species')

# Saving the output result in PNG and store in images folder
plt.savefig("images\histogram.png")
```

```python
# Ratio between length and width of Sepal and Petal
sepRatio = irisData['SepalLength_Cm'] / irisData['SepalWidth_Cm']
petRatio = irisData['PetalLength_Cm'] / irisData['PetalWidth_Cm']

# multiple histograms from a long-form dataset with hue mapping:
sns.displot(irisData, x=sepRatio, hue='Species').set(title='Sepal Ratio')
plt.savefig("images\histogram0.png")
sns.displot(irisData, x=petRatio, hue='Species').set(title='Petal Ratio')
plt.savefig("images\histogram1.png")
# Saving the output result in PNG and store in images folder
plt.savefig("images\histogram2.png")
sns.displot(irisData, x=sepRatio, col='Species')
# Saving the output result in PNG and store in images folder
plt.savefig("images\histogram3.png")
sns.displot(irisData, x=sepRatio, hue='Species', kind='kde', fill='true')
# Saving the output result in PNG and store in images folder
plt.savefig("images\histogram4.png")

```

```python
sns.histplot(data = irisData, x = "SepalLength_Cm", y = "SepalWidth_Cm", hue="Species")
# Saving the output result in PNG and store in images folder
plt.savefig("images\hist_plot.png")
```

```python
# Add a kernel density estimate to smooth the histogram, providing complementary information about the shape of the distribution: kde = True
sns.histplot(data = irisData, x = "SepalLength_Cm", hue="Species", kde=True)
# Saving the output result in PNG and store in images folder
plt.savefig("images\hist_plot2.png")
```
